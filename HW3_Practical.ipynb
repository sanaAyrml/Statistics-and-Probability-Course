```r
makeWY <- function(datas){
    y<- matrix(0,nrow = 51,ncol = 100000)
    w<- matrix(0,nrow = 51,ncol = 100000)
    for(i in 0:50){
      e <- sample(1:100,100-i)
      y[i+1,]<- colSums(datas[e[1:50],])
      w[i+1,]<- colSums(datas[tail(e,50),])
    }
    return(list(y,w))
}
```

```r
wy <- makeWY(datas)
y <- wy[[1]]
w <- wy[[2]]
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <ul>
                <li>                در انتهای مرحله‌ي قبل باید به 50 بردار 100000تایی برای $Y$ و 50 داده برای $W$ رسیده باشید. برای هر جفت‌های متناظر آنها (یعنی $W$ و $Y$ ای که برای هر $p$ به دست آورده‌اید) مقدار $cor(W, Y)$ را بدست آورید و به همراه $p$ متناظر در دیتافریمی قرار دهید.                 </li>
        </ul>

```r
makeDataFrame <- function(w,y){
    dataframe <- data.frame(p = c(0:50), cor = 0)
    for(i in 1:51){
      dataframe$cor[i] = cor(w[i,], y[i,])
    }
    return(dataframe)
}
```

```r
dataframe <- makeDataFrame(w,y)
```

میخواهیم داخل دایره‌ای به شعاع $R$ با توزیع یکنواخت نقطه‌ای انتخاب کنیم. به عبارتی احتمال اینکه نقطه‌‌ در محیطی داخل دایره به مساحت $S$ قرار بگیرد برابر با $\frac{S}{\pi R^2}$ باشد. 
</ul>
        <ul>
       الف)
       دو روش مختلف زیر را در $R$ شبیه‌سازی کنید و عملکرد هر یک را بررسی کنید.
        </ul>
        <ul>
        <li>
        اگر $X$ و $Y$ مختصات کارتزین باشند و 
        </li>
        </ul>
        <div dir = "ltr">
        $
  \begin{cases}
X \sim Uniform(-R, R)\\
Y \sim Uniform({\sqrt{R - X^2}, -\sqrt{R - X ^ 2})}
  \end{cases}
$
        </div>
    </font>
</div>

<div dir = "rtl" style="direction:rtl;line-height:300%;">

```r
X <- runif(10000,-10,10)
Y <- array(0.0,10000)
for(i in 1:10000){
  Y[i] <- runif(1,-(sqrt(100-(X[i]**2))),sqrt(100-(X[i]**2)))
}
n <- 0
samples <- c(1:10)
p <- c(0,10)
for(j in 1:10){
  for(i in 1:10000){
   if(X[i]**2 + Y[i]**2 <= samples[j]**2){
      n <- n+1
   }
  }
  p[j]<-n/10000
  n <- 0
}
print(p)
dataframe <- data.frame(x=(samples/10)**2,y=p)
ggplot(data = dataframe) + 
  geom_line(mapping = aes(x=x,y = y))
```

\begin{cases}

\theta \sim Uniform(0, ~2\pi)

```r
r <- runif(10000,0,10)
teta <- runif(10000,0,2*pi)
n <- 0
samples <- c(1:10)
p <- c(0,10)
for(j in 1:10){
  for(i in 1:10000){
    if((r[i]*cos(teta[i]))**2 +(r[i]*sin(teta[i]))**2 <= samples[j]**2){
      n <- n+1
    }
  }
  p[j]<-n/10000
  n <- 0
}
print(p)
dataframe <- data.frame(x=(samples/10)**2,y=p)
ggplot(data = dataframe) + 
  geom_line(mapping = aes(x=x,y = y))
```
